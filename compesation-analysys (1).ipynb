```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

### Merging two datasets

```python
df1 = pd.read_csv('/Users/nikolina/Documents/D 4.2/Data 4.2/Compesations in IT Industry 2019.csv')
df2 = pd.read_csv('/Users/nikolina/Documents/D 4.2/Data 4.2/Compesations in IT Industry 2020.csv')
```

```python
df2.head()
```

```python
df1.info()
```

```python
df2.info()
```

```python
df1.columns = df1.columns.str.strip()
df2.columns = df2.columns.str.strip()
```

```python
df1 = df1.rename(columns={'Zeitstempel': 'Timestamp', 'Position (without seniority)': 'Position', 'Yearly brutto salary (without bonus and stocks)': 'Yearly brutto salary (without bonus and stocks) in EUR'})
df2 = df2.rename(columns={'Total years of experience': 'Years of experience',  'Annual brutto salary (without bonus and stocks) one year ago. Only answer if staying in the same country': 'Yearly brutto salary (without bonus and stocks) one year ago. Only answer if staying in same country'})
df1['Timestamp'] = pd.to_datetime(df1['Timestamp'], format='%d.%m.%Y %H:%M:%S')
df2['Timestamp'] = pd.to_datetime(df2['Timestamp'], format='mixed', dayfirst=True)
```

```python
df1['Yearly bonus + stocks in EUR'] = df1['Yearly bonus'] + df1['Yearly stocks']
df1['Annual bonus+stocks one year ago. Only answer if staying in same country'] = df1['Yearly bonus one year ago. Only answer if staying in same country'] + df1['Yearly stocks one year ago. Only answer if staying in same country']
```

```python
df1.drop(columns=['Number of home office days per month', 'Company name', '0', 'Company business sector', 'Yearly bonus', 'Yearly stocks', 'Yearly bonus one year ago. Only answer if staying in same country', 'Yearly stocks one year ago. Only answer if staying in same country'], inplace=True)
df2.drop(columns=['Years of experience in Germany', 'Other technologies/programming languages you use often', 'Employment status', 'Have you lost your job due to the coronavirus outbreak?', 'Have you been forced to have a shorter working week (Kurzarbeit)? If yes, how many hours per week', 'Have you received additional monetary support from your employer due to Work From Home? If yes, how much in 2020 in EUR'], inplace=True)
```

```python
df2['Years of experience'].unique()
```

```python
df2['Years of experience'] = df2['Years of experience'].replace({'1,5': '1.5', '2,5': '2.5', '6 (not as a data scientist, but as a lab scientist)': '6', '15, thereof 8 as CTO': '15', '1 (as QA Engineer) / 11 in total': '1', 'less than year': '0.5'})
df2['Years of experience'] = df2['Years of experience'].astype(float)
df1['Years of experience'] = df1['Years of experience'].astype(float)
```

```python
df2['Number of vacation days'].unique()
```

```python
df2['Number of vacation days'] = df2['Number of vacation days'].apply(pd.to_numeric, errors='coerce')
df2['Yearly bonus + stocks in EUR'] = df2['Yearly bonus + stocks in EUR'].apply(pd.to_numeric, errors='coerce')
df2['Annual bonus+stocks one year ago. Only answer if staying in same country'] = df2['Annual bonus+stocks one year ago. Only answer if staying in same country'].apply(pd.to_numeric, errors='coerce')
```

```python
df = pd.merge(df1, df2, how='outer')
```

```python
pol = df['Gender'].value_counts()

df = df[df['Gender'] != 'Diverse']
```

```python
senioritet = df['Seniority level'].value_counts()

ostalo = senioritet[senioritet < 2].sum()
senioritet['Other'] = ostalo
senioritet = senioritet[senioritet >= 2].sort_values()
```

```python
df['Years of experience'].describe()
```

```python
df.loc[df['Years of experience'] > 50, 'Years of experience'] = np.nan
```

```python
pozicija = df['Position'].value_counts()

ostalo = pozicija[pozicija < 10].sum()
pozicija['Other'] = ostalo
pozicija = pozicija[pozicija >= 10].sort_values()
```

```python
df['Company size'].value_counts()
```

```python
df['Company size'] = df['Company size'].replace({'101-1000': '100-1000', '51-100': '50-100', 'Nov-50': '10-50', 'Oct-50': '10-50'})
```

```python
df['Yearly brutto salary (without bonus and stocks) in EUR'].max()
```

```python
df['Yearly brutto salary (without bonus and stocks) one year ago'].max()
```

```python
df.loc[df['Yearly brutto salary (without bonus and stocks) in EUR'] > 1_000_000, 'Yearly brutto salary (without bonus and stocks) in EUR'] = np.nan
df.loc[df['Yearly brutto salary (without bonus and stocks) one year ago'] > 1_000_000, 'Yearly brutto salary (without bonus and stocks) one year ago'] = np.nan
```

```python
df['Number of vacation days'].max()
df.loc[df['Number of vacation days'] > 80, 'Number of vacation days'] = np.nan
```

### Analysys

```python
pol_godine = df.groupby('Gender')['Age'].mean()

plt.bar(pol_godine.index, pol_godine.values, edgecolor='black', color='skyblue')
plt.title('Average age by gender');
```

```python
plata_pol = df.groupby('Gender')['Yearly brutto salary (without bonus and stocks) in EUR'].mean()

plt.bar(plata_pol.index, plata_pol.values, edgecolor='black', color='skyblue')
plt.title('Average salary by gender');
```

```python
df_poz = df[df['Position'].isin(pozicija.index)]
plate_po_poz = df_poz.groupby('Position')['Yearly brutto salary (without bonus and stocks) in EUR'].mean().sort_values()

plt.barh(plate_po_poz.index, plate_po_poz.values, edgecolor='black', color='skyblue')
plt.title('Average salary by position');
```

```python
iskustvo_plate = df.groupby('Years of experience')['Yearly brutto salary (without bonus and stocks) in EUR'].mean()

plt.scatter(iskustvo_plate.index, iskustvo_plate.values)
plt.title('Relationship between Years of experience and Yearly brutto salary');
```

```python
df['Year'] = df['Timestamp'].dt.year

god_plata = df.groupby('Year')['Yearly brutto salary (without bonus and stocks) in EUR'].mean()

god_plata.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Yearly brutto salary by year');
```

```python
df['Salary change'] = abs(df['Yearly brutto salary (without bonus and stocks) in EUR'] - df['Yearly brutto salary (without bonus and stocks) one year ago'])

promena_plate = df.groupby('Gender')['Salary change'].mean()

plt.bar(promena_plate.index, promena_plate.values, color='skyblue', edgecolor='black')
plt.title('Yearly salary change by gender');
```
